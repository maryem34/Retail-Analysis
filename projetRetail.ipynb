# Task 1 : Load the Data

```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
import plotly.express as px
```

```python
# 1. Load the dataset
excel_file_path = '/Online Retail.xlsx'

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(excel_file_path)
df.head() # Display first few rows
```

```python
# Display information about the dataFrame (Check data types and missing values)
df.info()
# Display summary statistics
df.describe(include="all")
```

```python
# Drop NAN
df.dropna(inplace=True)  # Drop rows with missing values
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])  # Convert date column to datetime
```

```python
df['CustomerID'].fillna('Unknown', inplace=True)
customer_data= df.groupby('CustomerID').agg({'InvoiceNo':'nunique',
                                             'StockCode': 'nunique',
                                             'Description': 'nunique',
                                             'Quantity': 'sum',
                                             'UnitPrice':'mean',
                                             'Country': 'first',
                                             'InvoiceDate': ['min','max'],

                                             }).reset_index()
customer_data.columns = ['CustomerID','numTransac','numuniqueProducts','NumUDesc','Total_quan','avg','country','firstInvoiceDate','lastInvoiceDate']
print(customer_data)
```

```python
# Create a new column to categorize special codes
df['SpecialCodeCategory'] = 'Regular'
df.loc[df['StockCode'].isin(['POST', 'D', 'M', 'PADS', 'DOT', 'CRUK', 'C2', 'm', 'B']), 'SpecialCodeCategory'] = 'Special'
df
```

```python
# "B" means Adjust bad depdt we have 3
# "m" means manual we have only one
# 'POST', 'D', 'M', 'PADS', 'DOT', 'CRUK','C2','m',"B" we have 2782 rows  8 columns" '76 with new dataFrame'
# List of codes to exclude
exclude_codes = ['POST', 'D', 'M', 'PADS', 'DOT', 'CRUK','C2','m','B']
new = df.dropna()
# Create a condition to filter out rows with specified codes
condition = new['StockCode'].isin(exclude_codes)

# Apply the filter
filtered_entries = new[condition]
filtered_entries
```

# Task 2 : Explore the Data

```python
# Extract year, month, day, and hour
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['Hour'] = df['InvoiceDate'].dt.hour

# Display the updated DataFrame
df.head()
```

```python
# Visual exploration xith plots "Monthly sales"
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Month'] = df['InvoiceDate'].dt.month
monthly_sales = df.groupby('Month')['Quantity'].sum()
plt.bar(monthly_sales.index, monthly_sales)
plt.xlabel('Month')
plt.ylabel('Total Quantity Sold')
plt.title('Monthly Sales')
plt.show()
```

```python
# Monthly sales trend
monthly_sales_trend = df.groupby(['Year', 'Month'])['Quantity'].sum()
monthly_sales_trend.plot(kind='bar', figsize=(12, 6))
plt.xlabel('Month')
plt.ylabel('Total Quantity Sold')
plt.title('Monthly Sales Trend')
plt.show()
```

```python
# Hourly sales trend (Transaction)
hourly_sales_trend = df.groupby(['Hour'])['Quantity'].sum()
hourly_sales_trend.plot(kind='bar', figsize=(12, 6))
plt.xlabel('Hours')
plt.ylabel('Total Quantity Sold')
plt.title('Hourly Sales Trend')
plt.show()
```

1.  *** Quantity vs. UnitPrice:*** The correlation coefficient is close to zero (-0.001235), indicating a very weak negative correlation between the quantity of products sold and the unit price. 2.  *** Quantity vs. Month: ***The correlation coefficient is close to zero again (-0.001116), indicating a very weak negative correlation between the quantity of products sold and the month in which the transactions occurred. This suggests that the month of the transaction does not have a strong linear relationship with the quantity of products sold.
1.   UnitPrice vs. Month: The correlation coefficient is close to zero as well (-0.000497), indicating a very weak negative correlation between the unit price of products and the month in which the transactions occurred. This suggests that the unit price is not strongly correlated with the month of the transaction.

```python
# Analyzing sales by country it shows that UK has the highest sales
country_sales = df.groupby('Country')['Quantity'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
country_sales.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Total Quantity Sold')
plt.title('Sales by Country')
plt.show()
```

```python
# Exclude the UK for better visualization
top_countries = country_sales[1:11]  # Exclude the UK, display top 10 countries

# Visualize top countries
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('Total Quantity Sold')
plt.title('Top 10 Countries (excluding the UK) - Sales')
plt.show()
```

```python
# Monthly sales trend for top countries
top_countries_list = top_countries.index.tolist()
top_countries_trend = df[df['Country'].isin(top_countries_list)]
monthly_trend_by_country = top_countries_trend.groupby(['Country', df['InvoiceDate'].dt.month])['Quantity'].sum().unstack()

plt.figure(figsize=(12, 6))
monthly_trend_by_country.plot(kind='bar', stacked=True)
plt.xlabel('Month')
plt.ylabel('Total Quantity Sold')
plt.title('Monthly Sales Trend for Top Countries')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Sales analysis for a specific country (e.g., France)
sales_in_france = df[df['Country'] == 'France']
sales_in_france
```

```python
# Identify the busiest months in terms of sales_in_france
busiest_months = df.groupby(df['InvoiceDate'].dt.to_period("M")).size().sort_values(ascending=False)
busiest_day = df.groupby(df['InvoiceDate'].dt.day_name()).size().sort_values(ascending=False)
top_selling = df.groupby(['Description','StockCode']).agg({'Quantity':'sum'}).sort_values(by='Quantity',ascending=False)
#Identify outliers using Z_score method
z_sc = (df['Quantity']- df['Quantity'].mean())/df['Quantity'].std()
outliers = df[abs(z_sc)>3]
#Display the results
print("Busiest months")
print(busiest_months)
print("Busiest Days of the weeks")
print(busiest_day)
print("THE TOP SELLING PRODUCTS BASED ON QUANTITY SOLD")
print(top_selling)
print("OUTLIERS OR ANOMALIES")
print(outliers)
```

```python
# Check for negative or very low values in UnitPrice
negative_prices = df[df['UnitPrice'] < 0]
low_prices = df[(df['UnitPrice'] >= 0) & (df['UnitPrice'] < 1)]

# Display entries with negative or low prices
print("Entries with negative prices:")
print(negative_prices)

print("\nEntries with low prices (0 <= UnitPrice < 1):")
print(low_prices)
```
